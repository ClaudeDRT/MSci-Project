```python
import numpy as np
from scipy import signal
import math
import matplotlib.pyplot as plt
import scipy.io
from scipy.integrate import simps, trapz
```

Extracting Simulated Echoes from Jack Landy's Code

```python
def echo_angles_comparison(angle): #input angles are 0, 10, 20, 30, 40, 50, 60 , 70, 80 and 90
    angle_index = int(angle/10)
    sim_echo = scipy.io.loadmat('/Users/claudederijkethomas/Desktop/Ice/Proj/FEM_Sim_Angles.mat')
    simulated_echo = [sim_echo['P_t_ml_range'][1][angle_index][1][0][i][0] for i in range(len(sim_echo['P_t_ml_range'][1][angle_index][1][0]))]
    simulated_echo_area = trapz(simulated_echo, dx = 1)
    norm_sim_echo = [x/simulated_echo_area for x in simulated_echo]
    return norm_sim_echo

plt.title('Normalised Simulated Echoes at Different Angles', fontsize = 15)
plt.plot(echo_angles_comparison(0), label = '0°')
plt.plot(echo_angles_comparison(10), label = '10°')
plt.plot(echo_angles_comparison(20), label = '20°')
plt.plot(echo_angles_comparison(30), label = '30°')
plt.plot(echo_angles_comparison(40), label = '40°')
plt.plot(echo_angles_comparison(50), label = '50°')
plt.plot(echo_angles_comparison(60), label = '60°')
plt.plot(echo_angles_comparison(70), label = '70°')
plt.plot(echo_angles_comparison(80), label = '80°')
plt.plot(echo_angles_comparison(90), label = '90°')
plt.ylabel(r'Normalised Power $/W$', fontsize = 13)
plt.xlabel(r'range bin number', fontsize = 13)
plt.legend(loc = 'best')
plt.savefig('NormAnglesDoff500Lw200sigma_si0.002.png', dpi=200)
```
